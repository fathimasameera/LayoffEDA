## Reading the data

```python
df=pd.read_csv("D:/DATASCIENCE/EDA/layoffs_data.csv")
df
```

The dataset provides information on layoffs in different companies. It includes the following columns:

* Company: The name of the company where the layoffs occurred.
* Location_HQ: The headquarters location of the company.
* Industry: The industry or sector to which the company belongs.
* Laid_Off_Count: The number of employees laid off by the company.
* Percentage: The percentage of the company's workforce that was laid off.
* Date: The date when the layoffs took place.
* Funds_Raised: Information about the amount of funds raised by the company.
* Stage: The stage of development or growth of the company.
* Country: The country where the company is located.

## Data Cleaning

```python
# Dropping off the Un-Necessary columns from dataset

df1 = df.drop(['Source','Date_Added','List_of_Employees_Laid_Off'], axis=1)
df1
```

```python
# Fetching the duplicated row count in the dataset 

df1.duplicated().sum()
```

```python
#Finding the duplicated row

df1[df1.duplicated()]
```

```python
# Removing duplicate values from the dataset

df1.drop_duplicates(inplace = True)
```

```python
# Finding number of missing values by column

df1.isnull().sum()
```

```python
# Dropping off the null values in the dataset
 
df1.dropna(inplace = True)
```

```python
#To count the number of rows of each columns

df1.count()
```

```python
# Renaming columns

df1.rename(columns={'Percentage': 'Laid_Off_Percentage', 'Date': 'Laid_Off_Date'}, inplace=True)
```

```python
df1.info()
```

```python
# Since the count of people laid off is a whole number and typically doesn't have fractional values, it makes sense to represent it as an integer for simplicity and clarity.
# Changing data types

df1[['Laid_Off_Count']]=df1[['Laid_Off_Count']].astype(int)
```

## Funds Raised vs Industry and Company

```python

# Grouping by Industry and Company and aggregating sum of and funds raised
industry_summary = df1.groupby(['Industry', 'Company']).agg({
    'Funds_Raised': 'sum'
})
industry_summary_sorted = industry_summary.sort_values(by='Funds_Raised', ascending=False)
industry_summary_sorted
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=industry_summary_sorted.head(20), x='Company', y='Funds_Raised', hue='Industry', palette='Set1', s=100)
plt.title('Total Funds Raised by Companies in Different Industries')
plt.xlabel('Company')
plt.ylabel('Total Funds Raised')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Industry', bbox_to_anchor=(1, 1))
plt.show()
```

## Conclusion & Interpretation from Data
### 1.Retail industry experienced the highest number of layoffs, while manufacturing industry had the least.
### 2.Finance industry has the highest laid off percentage.
### 3.The retail industry emerged as the most impacted sector, with the highest count of job losses, while the crypto industry, although among the top ten, positioned itself at the tenth spot in the list.
### 4.Among the top five giants laying off employees, Amazon took the lead, followed by Google and Meta, showing significant restructuring within the tech industry's major players.
### 5.In 2022, a significant peak in layoffs was observed, and this trend continued to increase in 2023, as demonstrated by the graphical representation.
### 6.In January, the highest number of layoffs were recorded, which aligns with the common practice of companies conducting financial year closing and settlements during this period.
### 7.The United States of America reported the highest number of layoffs, exceeding 0.2 million, while India recorded the second-highest count, with less than 50,000 layoffs.
### 8.The San Francisco Bay Area had the highest number of companies with maximum layoffs.
### 9.Most of the companies are in the Post-IPO stage and four of those companies stands top in the layoff count.
### 10.Media industry is leading in fund raising. Furthermore, within the Media industry, Netflix stands out as the top company in terms of total funds raised while Uber in the Transportation industry grabs the second spot.
